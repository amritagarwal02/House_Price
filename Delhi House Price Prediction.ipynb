```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("MagicBricks.csv")
```

```python
df2=df.drop(["Type","Transaction","Status","Furnishing"],axis="columns")
```

```python
df2
```

```python
df2.Per_Sqft=df2.Price/df2.Area
```

```python
df2
```

```python
df2.Per_Sqft=df2.Per_Sqft.round(0)
```

```python
df2
```

```python
df2["Locality"]=df2["Locality"].apply(lambda x: x.strip())
```

```python
df2.Locality.value_counts()
```

```python
df2.isnull().sum()
```

```python
df3=df2
```

```python
split_locality=df3.Locality.str.split(",",expand=True)
split_locality.rename(columns={0:"First",1:"Second",2:"Third"},inplace=True)
split_locality
```

```python
def sep(x):
    if(str(x.Third)=="None"):
        if(str(x.Second)=="None"):
            return str(x.First)
        else:
            return str(x.Second)
    else:
        return str(x.First)
```

```python
split_locality["Final"]=split_locality.apply(sep,axis=1)
```

```python
split_locality
```

```python
split_locality=split_locality["Final"]
split_locality
```

```python
df4=pd.concat([df3,split_locality],axis=1)
```

```python
df4
```

```python
df5=df4.drop(["Locality"],axis="columns")
```

```python
df5
```

```python
df5.rename(columns={"Final":"Locality"},inplace=True)
```

```python
df5
```

```python
len(df5.Locality.value_counts()<2)
```

```python
df5["Locality"]=df5["Locality"].apply(lambda x: x.strip())
```

```python
df5.Locality.value_counts()
```

```python
location_stats=df5.groupby("Locality")["Locality"].agg("count").sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<10])
```

```python
location_stats_less_than_10=location_stats[location_stats<10]
location_stats_less_than_10
```

```python
df6=df5
```

```python
df6["Locality"]=df6.Locality.apply(lambda x: "other" if x in location_stats_less_than_10 else x)
df6
```

```python
len(df6.Locality.unique())
```

```python
len(df6[df6.Area/df6.BHK<300])
```

```python
df6.isnull().sum()
df6=df6.dropna()
```

```python
df7=df6
df7
```

```python
indexArea=df7[df7.Area/df7.BHK<300].index
df7.drop(indexArea,inplace=True)
```

```python
df7=df7.drop(["Per_Sqft"],axis="columns")
```

```python
dummies=pd.get_dummies(df7.Locality)
dummies.head(3)
```

```python
df8=pd.concat([df7.drop(["Locality"],axis="columns"),dummies],axis="columns")
```

```python
df8
```

```python
df8=df8.drop(["other"],axis="columns")
```

```python
df8
```

```python
X=df8.drop("Price",axis="columns")
X
```

```python
Y=df8.Price
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf=LinearRegression()
lr_clf.fit(X_train,Y_train)
lr_clf.score(X_test,Y_test)
```
